# Climate Downscaling:

In this notebook, an essential component for making climate projections is addressed with spatial and non-spatial machine learning methods, namely climate downscaling Downscaling is calibrating a global climate model with spatially-limited local temperature measurements.

For climate models, calibration can be done in various ways. One way of performing calibration is the so-called statistical downscaling that forms a statistical relationship between global climate models and temperature observations. For the inherently spatial problem of downscaling, we evaluate regression methods based on their method of handling spatial information.

1. Non-spatial (generic) machine learning
2. Spatial machine learning
3. Non-spatial machine learning with geoenriched predictors

```python

def visualizeFeatureClassBySEDF(myMap, fc, renderer_type, method, class_count, col, colors,alpha):
    sdf = pd.DataFrame.spatial.from_featureclass(fc)
    sdf.spatial.project(spatial_reference = {'wkid': 3857}) # change the spatial reference of the spatial enabled data frame same as the map
    sdf.spatial.plot(map_widget= myMap,renderer_type=renderer_type, method=method, class_count=class_count, col=col, colors=colors,alpha=alpha)
    
   
```

```python
zonalSurf.set_index('OBJECTID', inplace=True)
zonal850.set_index('OBJECTID', inplace=True)
zonal500.set_index('OBJECTID', inplace=True)
vortSurf.set_index('OBJECTID', inplace=True)
vort850.set_index('OBJECTID', inplace=True)
vort500.set_index('OBJECTID', inplace=True)
```

```python
## Combine all param
Allpram = pd.concat([zonalSurf, zonal850, zonal500,vortSurf, vort850, vort500], axis=1, join_axes=[zonalSurf.index])
```

```python
Allpram.head()
```

```python
# Drop the non-used columns
Allpram_new = Allpram.drop(['lon', 'lat','SHAPE'], axis=1)
```

```python
Allpram_new.head()
```

```python
# rename the columns
Allpram_new.columns =['zonalSurf', 'zonal850', 'zonal500', 'vortSurf', 'vort850', 'vort500']
```

## Explore the correlation between different explanatory variables 

```python
# Calculate the correlation matrix
Allpram_new.corr()
```

```python
# Map the scatter plot
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(Allpram_new)
```

```python
# Map the correlation matrix

import matplotlib.cm as cm

corr = Allpram_new.corr()
size = 7
fig, ax = plt.subplots(figsize=(size, size))
cax = ax.matshow(corr,cmap=cm.get_cmap('coolwarm'), vmin=-1,vmax=1)
fig.colorbar(cax, ticks=[1,0,-1], shrink=0.8)
plt.xticks(range(len(Allpram_new.columns)), Allpram_new.columns, rotation='vertical', fontsize=8)
plt.yticks(range(len(Allpram_new.columns)), Allpram_new.columns)

plt.show()
```
